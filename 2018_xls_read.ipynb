```python
nyc_concat_2018 = pd.concat([man_2018, bronx_2018, brooklyn_2018, queens_2018, statenisland_2018])
# nyc_concat_2018.head()
nyc_concat_2018.dtypes
```

```python
rolling_2018 = nyc_concat_2018.drop(columns=['EASE-MENT\n'])
# rolling_2018.head()
rolling_2018.dtypes
```

```python
rolling_2018.columns = rolling_2018.columns.str.strip().str.replace(' ', '_')
rolling_2018.dtypes
```

```python
new_column_names = {"TAX_CLASS_AS_OF_FINAL_ROLL_18/19" : "TAX_CLASS_AT_TIME_OF_SALE", 
                    "BUILDING_CLASS_AS_OF_FINAL_ROLL_18/19" : "BUILDING_CLASS_AT_TIME_OF_SALE"}
new_rolling_2018 = rolling_2018.rename(index=str, columns=new_column_names)
new_rolling_2018.dtypes
```

```python
fixed_boroughs_2018 = new_rolling_2018.replace({'BOROUGH': 
     {1: 'MANHATTAN', 2: 'BRONX', 3: 'BROOKLYN', 4: 'QUEENS', 5: 'STATEN ISLAND'}})
# fixed_boroughs_2018.head()
fixed_boroughs_2018['BOROUGH'].unique()
# fixed_boroughs_2018.dtypes
```
