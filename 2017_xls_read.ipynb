```python
destination2 = "XLS/2017/2017_bronx.xls"
bronx_2017 = pd.read_excel(destination2, skiprows=4)
bronx_2017.head()
```

```python
destination3 = "XLS/2017/2017_brooklyn.xls"
brooklyn_2017 = pd.read_excel(destination3, skiprows=4)
brooklyn_2017.head()
```

```python
brooklyn_2017.dtypes
```

```python
destination4 = "XLS/2017/2017_queens.xls"
queens_2017 = pd.read_excel(destination4, skiprows=4)
queens_2017.head()
```

```python
queens_2017.dtypes
```

```python
destination5 = "XLS/2017/2017_statenisland.xls"
statenisland_2017 = pd.read_excel(destination5, skiprows=4)
statenisland_2017.head()
```

```python
statenisland_2017.dtypes
```

```python
nyc_concat_2017 = pd.concat([man_2017, bronx_2017, brooklyn_2017, queens_2017, statenisland_2017])
# nyc_concat_2017.head()
nyc_concat_2017.dtypes
```

```python
rolling_2017 = nyc_concat_2017.drop(columns=['EASE-MENT\n'])
# rolling_2017.head()
rolling_2017.dtypes
```

```python
rolling_2017.columns = rolling_2017.columns.str.strip().str.replace(' ', '_')
rolling_2017.dtypes
```

```python
new_column_names = {"TAX_CLASS_AS_OF_FINAL_ROLL_17/18" : "TAX_CLASS_AT_TIME_OF_SALE", 
                    "BUILDING_CLASS_AS_OF_FINAL_ROLL_17/18" : "BUILDING_CLASS_AT_TIME_OF_SALE"}
new_rolling_2017 = rolling_2017.rename(index=str, columns=new_column_names)
new_rolling_2017.dtypes
```

```python
fixed_boroughs_2017 = new_rolling_2017.replace({'BOROUGH': 
     {1: 'MANHATTAN', 2: 'BRONX', 3: 'BROOKLYN', 4: 'QUEENS', 5: 'STATEN ISLAND'}})
# fixed_boroughs_2017.head()
fixed_boroughs_2017['BOROUGH'].unique()
# fixed_boroughs_2017.dtypes
```

```python
fixed_boroughs_2017.to_csv('nyc_rolling_2017.csv')
```
