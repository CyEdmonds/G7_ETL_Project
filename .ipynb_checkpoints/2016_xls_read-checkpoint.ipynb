```python
nyc_concat_2016 = pd.concat([man_2016, bronx_2016, brooklyn_2016, queens_2016, statenisland_2016])
# nyc_rolling_2016.head()
nyc_concat_2016.dtypes
```

```python
rolling_2016 = nyc_concat_2016.drop(columns=['EASE-MENT\n'])
# rolling_2016.head()
rolling_2016.dtypes
```

```python
rolling_2016.columns = rolling_2016.columns.str.strip().str.replace(' ', '_')
rolling_2016.dtypes
```

```python
new_column_names = {"TAX_CLASS_AT_PRESENT" : "TAX_CLASS_AT_TIME_OF_SALE", 
                    "BUILDING_CLASS_AT_PRESENT" : "BUILDING_CLASS_AT_TIME_OF_SALE"}
new_rolling_2016 = rolling_2016.rename(index=str, columns=new_column_names)
new_rolling_2016.dtypes
```

```python
fixed_boroughs_2016 = new_rolling_2016.replace({'BOROUGH': 
     {1: 'MANHATTAN', 2: 'BRONX', 3: 'BROOKLYN', 4: 'QUEENS', 5: 'STATEN ISLAND'}})
# fixed_boroughs_2016.head()
fixed_boroughs_2016['BOROUGH'].unique()
# fixed_boroughs_2016.dtypes
```
